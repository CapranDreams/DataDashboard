```python
import random

def generate_data(num_time_bins, num_frequency_bins):
    data = []
    for f in range(num_frequency_bins):
        rowVal = random.random()**2 * 100
        if rowVal < 50:
            rowVal = rowVal / 10
        if f < 10 and rowVal < 30:
            rowVal = 30
        for t in range(num_time_bins):
            cellVal = rowVal + random.random() * 10 - 5
            if cellVal < 0:
                cellVal = 0
            if cellVal > 100:
                cellVal = 100
            cellVal = int(cellVal)
            data.append(cellVal)
    return data

data = generate_data(num_time_bins = 180, num_frequency_bins = 100)
for x in data:
    print(x)

import matplotlib.pyplot as plt

plt.plot(range(len(data)), data)
plt.show()
```
